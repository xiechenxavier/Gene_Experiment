```python
import random

nbGene = 20
start = [x for x in range(0, nbGene)] # 0 -> 9
# an array
mat_classement = []
#function to generate a list
def generateLists(A,B,COUNT,NBligne,mat_classement):
    for i in range(NBligne):
        resultList = random.sample(range(A,B+1),COUNT)
        mat_classement.append(resultList)
```

```python
#1、利用递归生成
A=1; #最小随机数
B=20 #最大随机数
nbcol=20 #多少列
nbligne = 20
generateLists(A,B,nbcol,nbligne,mat_classement)
```

```python
len(mat_classement)
```

```python
from itertools import product

wij_table = []
for j in range(len(start)):
    wij_table.append([0 for i in range(len(start))])
def compterAllPairs():
    setstart = set(start)
    for ele in mat_classement:
       
        for i,j in product(setstart,setstart):
            if i != j :
                if(ele.index(i+1) < ele.index(j+1) ):
                    wij_table[i][j] += 1
    # calculer toutes les probabilites        
    for i,j in product(setstart,setstart):
        wij_table[i][j]/= len(mat_classement)


compterAllPairs()
wij_table         
        
```

```python
try:

    m = Model("mip1")
    # Create variables 变量是通过addVar()方法在一个模型对象上添加的
    # (或者addVars()，如果您希望一次添加多个变量)。变量总是与特定的模型相关联。
    x = m.addVar(vtype=GRB.BINARY, name="x")
    y = m.addVar(vtype=GRB.BINARY, name="y")
    z = m.addVar(vtype=GRB.BINARY, name="z")
    # Set objective   其中GRB.MAXIMIZE意义是最大化。
    # 请注意，虽然这个简单的示例使用显式的术语列表在一条语句中构建了目标，
    # 但是更复杂的程序通常会增量地构建它。例如
    # obj = LinExpr();obj += x;obj += y;obj += 2*z;model.setObjective(obj, GRB.MAXIMIZE);
    m.setObjective(x + y + 2 * z, GRB.MAXIMIZE)
    # Add constraint: x + 2 y + 3 z <= 4   添加约束
    m.addConstr(x + 2 * y + 3 * z <= 4, "c0")
    # Add constraint: x + y >= 1
    m.addConstr(x + y >= 1, "c1")

    m.optimize()
    for v in m.getVars():#m.getVars() 获取所有的变量
        print(v.varName, v.x) #一旦优化完成，我们就可以查询属性的值。
        #特别是，我们可以查询varName和x变量属性，以获得每个变量的名称和解决方案值
    print('Obj:', m.objVal) #我们也可以查询模型上的objVal属性，得到当前解决方案的目标值
except GurobiError:
        print('Error reported') #Python接口中的错误是通过Python异常机制处理的。在本例中，所有gurobi语句都包含在一个try块中，而except块将捕获所有相关错误。
```

```python
class Gurobi_Gene:
    
    def __init__(self):
        self.m = Model("GeneSolverModel")
        self.wij_table = []
        
    
    def importInstance(self,filename):
        '''
        cette fonction permet d'importer des donnes d'une fichier
        '''
        #la table de la probabilite d'i est avant d'j
        self.wij_table = (np.loadtxt(filename,delimiter=',')).tolist()
        #la taille est le nombre du Gene
        self.taille = len(self.wij_table)

        
    def parametrageModel(self):
        '''
        parametrer les contenues communes d'une configuration d'un model
        '''
        #les indices dans un ensemble
        setPosition = set([ele for ele in range(0,self.taille)])
        # (或者addVars()，如果您希望一次添加多个变量)。变量总是与特定的模型相关联。
        self.x = self.m.addVar(self.taille,self.taille,vtype=GRB.BINARY, name="x")
        # objectif function
        #  xsum(wij_table[i][j]*x[i][j] for i in setstart for j in setstart-{i} )
        objective = quicksum(self.x[i][j]*self.wij_table[i][j] for i,j in setPosition)
        self.m.setObjective(objective)
        
    # les contraintes principales contiennent constraint 1,2
    def constraintPrincipal(self):
        for i in range(0,len(self.taille)):
            for j in range(0,len(self.taille)):
                if(i!=j):
                    self.m.addConstr(x[i,j]+x[j,i] == 1)
                    
        for i in range(0,len(self.taille)):
            self.m.addConstr(x[i,i] <= 0)
    
    # Formulation1 : 1,2,4
    def Formulation1(self):
        #les contraintes principales 1,2
        self.constraintPrincipal()
        
        #varaibles y
        self.y = self.m.addVar(self.taille,self.taille,vtype=GRB.BINARY, name="y")
        
        #les indices dans un ensemble
        setPosition = set([ele for ele in range(0,self.taille)])
 
        self.m.addConstr((self.y[i,j] <= self.x[i,j] for j in setPosition-{i}) for i in setPosition)
        
    def PrintResult(self,Formulation):
        if(Formulation == 'Formulation1'):
            self.Formulation1()
        self.m.optimize()
        for v in self.m.getVars():#m.getVars() 获取所有的变量
            print(v.varName, v.x) #一旦优化完成，我们就可以查询属性的值。
            #特别是，我们可以查询varName和x变量属性，以获得每个变量的名称和解决方案值
        print('Obj:', self.m.objVal) #我们也可以查询模型上的objVal属性，得到当前解决方案的目标值
        
Gene_rank1 = Gurobi_Gene()
Gene_rank1.importInstance("./Gene/Gurobi/instances/Gene10_10.txt")
print(Gene_rank1.wij_table)
Gene_rank1.parametrageModel()
Gene_rank1.PrintResult("Formulation1")
```
